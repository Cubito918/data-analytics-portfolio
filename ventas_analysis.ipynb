```python
import pandas as pd

# Cargar dataset desde GitHub

url = "https://raw.githubusercontent.com/Cubito918/data-analytics-portfolio/refs/heads/main/ventas-analysis/ventas.csv"

df = pd.read_csv(url)
df.head()

# Total Vendido por Producto
ventas_producto = df.groupby("producto")["monto"].sum()
print(ventas_producto)

#Total Vendido por Mes
ventas_mes = df.groupby("fecha")["monto"].sum()
print(ventas_mes)

# Producto "TOP" (m치s vendido)
producto_top = ventas_producto.idxmax()
print(producto_top)

#Cliente "TOP" (el que m치s gast칩)
cliente_top = df.groupby("id_cliente")["monto"].sum().idxmax()
print(cliente_top)
```

- Abril registra el mayor ingreso mensual.
