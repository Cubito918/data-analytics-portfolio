```python
import pandas as pd

url = "https://raw.githubusercontent.com/Cubito918/data-analytics-portfolio/refs/heads/main/clientes-analysis/clientes.csv"

df = pd.read_csv(url)
df.head()


#cliente que más gastó
cliente_top = df.sort_values(by="total_gastado", ascending=False).iloc[0]
print(cliente_top)

#promedio de gasto
promedio = df["total_gastado"].mean()
print(promedio)

#método para segmentar clientes por nivel
def segmento(gasto):
    if gasto >= 10000:
        return "Alto valor"
    elif gasto >= 5000:
        return "Valor medio"
    else:
        return "Bajo valor"

df["segmento"] = df["total_gastado"].apply(segmento)
df

#conteo por segmento
df["segmento"].value_counts()
```

##  Insights

- El cliente 105 pertenece al segmento **Alto Valor**.
- La mayoría de clientes se concentra en el segmento **Valor Medio**.
- Se recomienda priorizar retención de clientes de alto valor.
