```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/Cubito918/data-analytics-portfolio/refs/heads/main/ventas-dashboard/ventas.csv"

df = pd.read_csv(url)
df.head()

# Ventas por producto (ahora mostrando grfico de barras)

ventas_producto = df.groupby("producto")["monto"].sum()

ventas_producto.plot(kind="bar", title="Ventas Totales por Producto")
plt.show()

# Ventas por mes (ahora mostrando grfico de lnea)

ventas_mes = df.groupby("fecha")["monto"].sum()

ventas_mes.plot(kind="line", title="Evolucin de Ventas por Mes")
plt.show()

# Producto destacado

producto_top = ventas_producto.idxmax()
print(producto_top)
```

##  Insights del Dashboard

- Laptop genera el mayor volumen de ingresos.
- Se observa una tendencia creciente hasta abril.
- Se recomienda invertir en campaas durante meses con mayor rendimiento.
